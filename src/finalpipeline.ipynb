```python
import pandas as pd
from processing import clean_data
from modelling import get_student_pipeline
import numpy as np
from sklearn.model_selection import train_test_split
from datetime import datetime
import os
import joblib

# Load the dataset 
DATA_PATH = r'..\data\raw\datasetOne.csv'
df = pd.read_csv(DATA_PATH)
```

```python
# Data cleaning from processing.py 
target_col = "Target"
df_clean = clean_data(df, target_col)
```

```python
# Train/Test split 

RANDOM_SEED = 42
TEST_SIZE = 0.2 

# Remove rows with 'Enrolled' (2.0)
df_binary = df_clean[df_clean[target_col] != 2.0].copy()

# Define features (X) and target (y) for the binary classification (Dropout vs Graduate)
X = df_binary.drop(columns=[target_col]) 
y = df_binary[target_col]

# Split the data into training and testing sets with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=TEST_SIZE, 
    random_state=RANDOM_SEED,
    stratify=y 
)

print(f"Training set size: {X_train.shape[0]} samples. Testing set size: {X_test.shape[0]} samples")
```

```python
# Define the features to be scaled
scale_features = [
   'Curricular units 1st sem (enrolled)', 
    'Curricular units 1st sem (evaluations)', 
    'Curricular units 1st sem (approved)', 
    'Curricular units 1st sem (grade)',
    'Academic_Efficiency_Ratio',
    'Total_Academic_Momentum',
    'Grade_Velocity_Trend',
    'Recovery_Signal'
]

# Create the pipeline
pipeline = get_student_pipeline(scale_features)

# Train & Store the pipeline
trained_model = pipeline.fit(X_train, y_train)
```

```python
# Save the entire pipeline using joblib

VERSION = "v1"  
TIMESTAMP = datetime.now().strftime("%Y%m%d_%H%M")
SAVE_DIR = '../models'
os.makedirs(SAVE_DIR, exist_ok=True)

# Define the filename with version and timestamp
filename = f"XGBoost_Pipeline_{VERSION}_{TIMESTAMP}.joblib"
full_path = os.path.join(SAVE_DIR, filename)

# Save the trained pipeline
joblib.dump(trained_model, full_path)

print(f"Saved Full Pipeline successfully: {filename}")
```
